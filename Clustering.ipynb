```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X,y_true=make_blobs(n_samples=500,centers=3,cluster_std=0.60,random_state=42)
```

```python
df=pd.DataFrame(X,columns=["Feature1","Feature2"])
```

```python
df.head()
```

```python
df.shape
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(df)
```

```python
inertia=[]
K_range=range(1,11)
```

```python
for k in K_range:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(K_range,inertia, marker='o')
```

```python
kmean_final=KMeans(n_clusters=3,random_state=42)
```

```python
cluser_labels=kmean_final.fit_predict(X_scaled)
```

```python
df.columns
```

```python
df['cluster']=cluser_labels
```

```python
sns.scatterplot(x=df["Feature1"],
                y=df["Feature2"],
                hue=df["cluster"],
                palette="viridis")
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,y_true=make_moons(n_samples=500,noise=0.05,random_state=42)
```

```python
df=pd.DataFrame(X,columns=["Feature1","Feature2"])
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(df)
```

```python
kmean_final=KMeans(n_clusters=2,random_state=42)
cluser_labels=kmean_final.fit_predict(X_scaled)
```

```python
df['kmeans_cluster']=cluser_labels
```

```python
sns.scatterplot(x=df["Feature1"],
                y=df["Feature2"],
                hue=df["kmeans_cluster"],
                palette="viridis")
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```
